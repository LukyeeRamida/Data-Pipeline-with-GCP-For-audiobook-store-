```python
import os

class Config:
  MYSQL_HOST = ''
  MYSQL_PORT = 3306
  MYSQL_USER = ''
  MYSQL_PASSWORD = ''
  MYSQL_DB =''
  MYSQL_CHARSET = ''
```

```python
import pymysql


connection = pymysql.connect(host=Config.MYSQL_HOST,
                             port=Config.MYSQL_PORT,
                             user=Config.MYSQL_USER,
                             password=Config.MYSQL_PASSWORD,
                             db=Config.MYSQL_DB,
                             charset=Config.MYSQL_CHARSET,
                             cursorclass=pymysql.cursors.DictCursor)
```

```python
with connection.cursor() as cursor:
  cursor.execute("show tables;")
  tables = cursor.fetchall()

  print(tables)
```

```python
with connection.cursor() as cursor:
  cursor.execute("SELECT * FROM audible_data;")
  result = cursor.fetchall()

  print(result)
```

```python
import pandas as pd
audible_data = pd.DataFrame(result).set_index("Book_ID")
```

```python
audible_data
```

```python
import sqlalchemy


engine = sqlalchemy.create_engine(
    "mysql+pymysql://{user}:{password}@{host}:{port}/{db}".format(
        user=Config.MYSQL_USER,
        password=Config.MYSQL_PASSWORD,
        host=Config.MYSQL_HOST,
        port=Config.MYSQL_PORT,
        db=Config.MYSQL_DB,
    )
)

sql = "SELECT * FROM audible_transaction"
audible_transaction = pd.read_sql_query(sql, engine)
audible_transaction
```

# Join audible_data and audible_transaction

```python
transaction = audible_transaction.merge(audible_data, how="left", left_on="book_id", right_on="Book_ID")
```

```python
transaction
```

# Get data from REST API


1.   Import requests package and pull data (Currency conversion rate) from REST API
2.   Convert data to pandas



```python
import requests

url = "https://r2de2-workshop-vmftiryt6q-ts.a.run.app/usd_thb_conversion_rate"
r = requests.get(url)

print(r.status_code)
result_conversion_rate = r.json()
```

```python
print(result_conversion_rate)
```

```python
conversion_rate = pd.DataFrame(result_conversion_rate).reset_index().rename(columns={"index": "date"})
conversion_rate[0:3]
```

# Join transaction and conversion_rate


*   Correct price column
*   Create new price column in THB



```python
transaction['date'] = transaction['timestamp']

transaction['date'] = pd.to_datetime(transaction['date']).dt.date
conversion_rate['date'] = pd.to_datetime(conversion_rate['date']).dt.date

final_df = transaction.merge(conversion_rate, how="left", left_on="date", right_on="date")
final_df
```

```python
final_df["Price"] = final_df.apply(lambda x: x["Price"].replace("$",""), axis=1)
final_df["Price"] = final_df["Price"].astype(float)
```

```python
final_df["THBPrice"] = final_df["Price"] * final_df["conversion_rate"]
final_df = final_df.drop('date', axis = 1)
final_df
```

```python
# Save to CSV
final_df.to_csv("output.csv", index=False)
```
